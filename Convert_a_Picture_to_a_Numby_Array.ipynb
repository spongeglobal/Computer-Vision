```python
import numpy as np
from PIL import Image

# Open the image file
img = Image.open('face.jpg')

# Convert the image to a NumPy array
img_array = np.array(img)

print(f"Image successfully converted to a NumPy array with shape: {img_array.shape}")
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_array)
plt.axis('off') # Hide axes ticks and labels
plt.show()
```

Let's analyze the pixel values in the `img_array`.

```python
print(f"Data type of the image array: {img_array.dtype}")
print(f"Shape of the image array: {img_array.shape}")
print(f"Minimum pixel value: {np.min(img_array)}")
print(f"Maximum pixel value: {np.max(img_array)}")
print(f"Mean pixel value: {np.mean(img_array):.2f}")
print(f"Standard deviation of pixel values: {np.std(img_array):.2f}")
```
